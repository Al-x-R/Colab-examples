```
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/an-2-an/data/master/cur01092019.csv')
df['month'] = pd.DatetimeIndex(df['exchangedate']).month
df['year'] = pd.DatetimeIndex(df['exchangedate']).year
df['day'] = pd.DatetimeIndex(df['exchangedate']).day
df.head(10)
```

```
newDf =  df[df['txt']=='Платина']
newDf.head(10)
```

```

newDf[ newDf ['txt']=='Платина'] [['txt','rate','exchangedate','month','year']].head(10)
```

```
newDf.plot(x = 'exchangedate', y = 'rate');
```

```
newDf.plot(x = 'exchangedate', y = 'rate', figsize=(15, 10), color='c', style='--');
```

```
newDf[newDf['year']==2019].pivot_table(values='rate', index=['year','month'], aggfunc='sum')
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['month'], aggfunc='sum').plot(figsize=(10, 6), color='c', style='--');
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['month','day'], aggfunc='sum').head(10)
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['month','day'], aggfunc='sum').plot(figsize=(15, 10), color='c', style='--');
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['month'], aggfunc='sum').plot.pie(figsize=(10, 6),subplots=True);
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['year','month','day'], aggfunc='sum').head(10)
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['year','month'], aggfunc='mean')
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['month'], aggfunc='mean').plot.bar(figsize=(10, 6));
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['year','month'], aggfunc='max')
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['month'], aggfunc='max').plot.bar(figsize=(10, 6));
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['year','month'], aggfunc='min')
```

```
newDf[newDf['year']==2018].pivot_table(values='rate', index=['month'], aggfunc='min').plot.bar(figsize=(10, 6));
```

```
newDf[newDf['year']==2018].groupby('day')[['rate', 'month']].sum().sort_values(by='day')
```

```
newDf[newDf['year']==2018].pivot_table(values=['rate','day'], index='month', columns='txt', aggfunc='sum')
```
